## Reading all results

```python
field = 'temperature'
model_id = 'vbynozry'

results_path = f'/p/home/jusers/semcheddine1/juwels/dev_atmorep/atmorep/results/id{model_id}/results_id{model_id}_epoch00000_target.zarr'

store_target = zarr.ZipStore(results_path)
ds_target = zarr.group( store=store_target)

i = 0

# ml = 96 # Todo: generalize for all levels

datetime_all = [ds_target[ f'{field}/{i_str}/datetime'][:] for i_str in ds_target[ f'{field}']]
datetime_all = np.unique(np.array(datetime_all))

ds_o = xr.Dataset( coords={ 'ml' : ds_target[ f'{field}/sample=00000/ml'][:], # Todo: generalize for all levels
                            'datetime': datetime_all, 
                            'lat' : np.linspace( -90., 90., num=180*4+1, endpoint=True), 
                            'lon' : np.linspace( 0., 360., num=360*4, endpoint=False) }
                             )

nlevels =  len(ds_target[ f'{field}/sample=00000/ml'][:])
time_frame = len(datetime_all)


ds_o['vo_target'] = (['ml', 'datetime', 'lat', 'lon'], np.zeros( (nlevels, time_frame, 721, 1440)))

for i_str in ds_target[ f'{field}']:
    ds_o['vo_target'].loc[ dict( datetime=ds_target[ f'{field}/{i_str}/datetime' ][:], 
        lat=ds_target[ f'{field}/{i_str}/lat' ][:],
        lon=ds_target[ f'{field}/{i_str}/lon' ][:]) ] = ds_target[ f'{field}/{i_str}/data' ] 
```

## Save plots

```python
vlevel_idx=0

vlevel = ds_o["ml"].values[vlevel_idx]
datetime_idx = 31
datetime = ds_o["datetime"].values[datetime_idx]

vmin = ds_o['vo_target'].values[vlevel_idx,datetime_idx,:,:].min()
vmax = ds_o['vo_target'].values[vlevel_idx,datetime_idx,:,:].max()


datatype = 'vo_target'
cmap = matplotlib.colormaps.get_cmap('RdBu_r')
fig = plt.figure( figsize=(10,5), dpi=300)
ax = plt.axes( projection=cartopy.crs.Robinson( central_longitude=0.))
ax.add_feature( cartopy.feature.COASTLINE, linewidth=0.5, edgecolor='k', alpha=0.5)
ax.set_global()
ax.set_title(f"Datatype = {datatype}, field = {field}, vlevel = {vlevel}, datetime = {datetime}", fontsize=10)
im = ax.imshow( np.flip(ds_o[datatype].values[vlevel_idx,datetime_idx,:,:], 0), 
          cmap=cmap, vmin=vmin, vmax=vmax,
          transform=cartopy.crs.PlateCarree( central_longitude=180.))
axins = inset_axes( ax, width="80%", height="5%", loc='lower center', borderpad=-2 )
fig.colorbar( im, cax=axins, orientation="horizontal")
datetime_str_abv = str(datetime)[:13]
plt.savefig(f'images/tempolar_interpolation_{datatype}_{field}_ml{vlevel}_date_{datetime_str_abv}.png')
# fig.show()
```
